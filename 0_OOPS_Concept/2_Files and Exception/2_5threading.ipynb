```python
"""
Executing python app.py command:

When you run the python app.py command, you are telling the Python interpreter (in this case, CPython) to execute the instructions written in the app.py file.
Compilation into machine code:

The Python interpreter compiles the human-readable Python code in app.py into machine code, which is a set of instructions that the computer's CPU (Central Processing Unit) can understand and execute.
Loading into memory (RAM):

Once the machine code is generated, the operating system (OS) is responsible for loading the program into the computer's memory (RAM). The RAM is a volatile form of memory that provides fast access to data and instructions needed for the program to run.
Moving instructions to the CPU:

After loading into memory, the OS moves the instructions of the program to the CPU. This transfer of instructions is typically done using buses, which are communication pathways between different components of the computer.
Execution in the CPU:

The CPU executes the instructions from the program. The OS manages this execution by coordinating the flow of instructions to the CPU.
Pipeline and Queue:

The OS often uses a queue or pipeline to manage the flow of instructions to the CPU. A queue is a data structure that follows the First In, First Out (FIFO) principle, where the instructions are lined up in the order they are received. The CPU then fetches instructions from the front of the queue and executes them.
Definition of Process and Thread:

A process is an instance of a program that is running on a computer. It includes the program's code, data, and system resources. Each process has its own memory space, and processes are generally isolated from each other.
A thread is a unit of execution within a process. Multiple threads can exist within a single process, and they share the same memory space. Threads within a process can run concurrently, allowing for more efficient use of resources.
In summary, when you execute a Python program, it goes through compilation, loading into memory, and execution by the CPU. The OS manages these processes and uses concepts like queues, processes, and threads to efficiently handle the execution of multiple programs on a computer.






"""
```

```python
url_list = ['https://raw.githubusercontent.com/itsfoss/text-files/master/agatha.txt' , 'https://raw.githubusercontent.com/itsfoss/text-files/master/sherlock.txt' ,'https://raw.githubusercontent.com/itsfoss/text-files/master/sample_log_file.txt' ]
```

```python
data_file_list = ['data1.txt', 'data2.txt','data3.txt']
```

```python
import urllib.request
def file_download(url,filename):
    urllib.request.urlretrieve(url, filename)
```

```python
thread1 = [threading.Thread(target=file_download, args=(url_list[i],data_file_list[i])) for i in range(len(url_list))]
```

```python
for t in thread1:
    t.start()
```
